```python
# Import the Abstract Syntax Trees module for safely evaluating Python literals
import ast

# Convert string representations of lists in the 'ingredient_names' column 
# into actual Python list objects using ast.literal_eval
df_full['ingred_list'] = df_full['ingredient_names'].apply(ast.literal_eval)
```

```python
df_full.iloc[0].ingred_list
```

```python
import pandas as pd
from collections import Counter

# N: How many top ingredients you want
N = 10

# To store results
results = []

# filter out categories with very few products
filtered_df = df_full.groupby('predicted_category').filter(lambda x: len(x) >= 10)


# Loop over each product_type group
for product_type, group in filtered_df.groupby('predicted_category'):
    total_products = len(group)
    
    # Flatten all ingredients for this group
    all_ingredients = sum(group['ingred_list'], [])
    
    # Count ingredient frequency
    counts = Counter(all_ingredients)
    
    # Get top N ingredients
    top_ingredients = counts.most_common(N)
    
    # Calculate percentage
    for ingred, count in top_ingredients:
        percentage = count / total_products * 100  # percentage
        results.append({
            'product_category': product_type,
            'ingredient': ingred,
            'count': count,
            'percentage': round(percentage, 2)
        })

# Create a result dataframe
top_ingredients_df = pd.DataFrame(results)

# Sort nicely
top_ingredients_df = top_ingredients_df.sort_values(['product_category', 'percentage'], ascending=[True, False])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import matplotlib.cm as cm
import matplotlib.colors as mcolors

# Count categories
category_counts = df_full['predicted_category'].value_counts()
category_counts = category_counts.sort_values()

# Normalize and create colors using Viridis
norm = mcolors.Normalize(vmin=category_counts.min(), vmax=category_counts.max())
cmap = cm.get_cmap('viridis')
colors = [cmap(norm(value)) for value in category_counts.values]

# Plot
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.barh(category_counts.index, category_counts.values, color=colors, edgecolor='black')

# Create ScalarMappable for colorbar
sm = cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])  # required to avoid warning

# Add colorbar to the figure
cbar = fig.colorbar(sm, ax=ax)
cbar.set_label('Number of Products')

# Labels and layout
ax.set_title('Product Category Distribution', fontsize=14)
ax.set_xlabel('Number of Products', fontsize=12)
ax.set_ylabel('Product Category', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Explode dataframe so each row is (product, ingredient)
df_exploded = df_full[['predicted_category', 'ingred_list']].explode('ingred_list')
```

```python
# Total number of products
total_products = df_full.shape[0]

# Ingredient appearance counts( number of products it appears in)
ingredient_counts = df_exploded.groupby('ingred_list')['predicted_category'].nunique() # Number of unique products

# Appearance percentage
ingredient_freq = df_exploded.groupby('ingred_list')['predicted_category'].count() / total_products * 100
ingredient_freq = ingredient_freq.rename('frequency_pct')
```

```python
import numpy as np 

# Create a crosstab : ingredient vs product type counts 
crosstab = pd.crosstab(df_exploded['ingred_list'], df_exploded['predicted_category'])

# Normalize by row to get p(i) distribution
crosstab_norm = crosstab.div(crosstab.sum(axis=1), axis=0)

# Calculate entropy each ingredient
entropy = -(crosstab_norm * np.log(crosstab_norm + 1e-9)).sum(axis=1) # smal epsilon to avoid log(0)
entropy = entropy.rename('entropy')
```

```python
ingredient_stats = pd.concat([ingredient_freq, entropy], axis=1).reset_index()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 8))
sns.scatterplot(
    x='entropy',
    y='frequency_pct',
    data=ingredient_stats
)
plt.xlabel('Entropy (spread across product types)')
plt.ylabel('Frequency (% of products)')
plt.title('Ingredient Frequency vs Entropy')
plt.grid(True)
plt.show()
```

```python
# Define N
N = 10

# Filter top N per product_type
topN_df = (top_ingredients_df.groupby('product_category')
           .head(N))

# Get all unique ingredients across all product types' Top-N
top_ingredients_all = set(topN_df['ingredient'])
print(f"Number of unique top ingredients: {len(top_ingredients_all)}")
```

```python
# Keep only ingredients that are in top_ingredients_all
df_full['ingred_list_top'] = df_full['ingred_list'].apply(lambda ingreds: [i for i in ingreds if i in top_ingredients_all])
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer(classes=list(top_ingredients_all))  # Fix the order
ingred_matrix = mlb.fit_transform(df_full['ingred_list_top'])

# Now, ingred_matrix shape: (num_products, num_top_ingredients)
print(ingred_matrix.shape)
```

```python
from sklearn.manifold import TSNE

# Initialize t-SNE
tsne = TSNE(n_components=2, random_state=42, perplexity=30)

# Fit and transform
tsne_results = tsne.fit_transform(ingred_matrix)

# Add back to DataFrame
df_full['tsne_1'] = tsne_results[:, 0]
df_full['tsne_2'] = tsne_results[:, 1]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 8))
sns.scatterplot(
    x='tsne_1', y='tsne_2',
    hue='predicted_category',
    palette='tab10',
    data=df_full,
    alpha=0.7
)
plt.title('t-SNE of Skincare Products by Top Ingredients')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
plt.show()
```
